```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
df_beta = pd.DataFrame([
    [1, 66.05],
    [2, 66.56],
    [4, 66.55],
    [6, 64.88],
    [8, 67.16],
    [10, 66.51],
    [12, 66.40],
    [14, 66.75],
    [16, 66.49],
    [18, 65.32],
    [20, 66.44],
    [22, 64.82],
    [24, 63.80],
    [26, 65.01],
], columns=["beta", "accuracy"])

df_beta.head()
```

```python
FONTSIZE = 20
fig, ax = plt.subplots(figsize=(10, 5))
df_beta.plot(
    x="beta", 
    y="accuracy", 
    ax=ax, 
    legend=None, 
    fontsize=FONTSIZE, 
    linewidth=5, 
    marker="o", 
    markersize=15
)
ax.set_xlabel(r"Coefficent $\beta$", fontsize=20)
ax.set_ylabel("Accuracy", fontsize=20)
fig.patch.set_facecolor("white")
```
